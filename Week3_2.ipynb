## Part 2: Data Summarization

```python
# Total "Gross" of all top 1000 movies
print(imdb['Gross'].sum())
```

```python
# Average "No_of_Votes" of all movies
print(imdb['No_of_Votes'].mean())
```

```python
# Count movies in each decade (e.g., ..., 1980, 1990, 2000, 2010, 2020)
# Sort decades by the number of movies in descending order
imdb['Decade'].value_counts().sort_values(ascending=False)
```

```python
# Count movies by different directors.
# Show the top 10 directors with the most movies in this list
imdb['Director'].value_counts().sort_values(ascending=False).head(10)
```

```python
# For each director, calculate the number of movies, average IMDB_Rating, and total gross.
# Ranked in descending order of total gross
# Show the top 10 directors with the most gross
imdb.groupby('Director').agg(
    Num_Movies=('Series_Title', 'count'),
    Avg_IMDB_Rating=('IMDB_Rating', 'mean'),
    Total_Gross=('Gross', 'sum')
).sort_values(by='Total_Gross', ascending=False).head(10)
```

```python
# Group movies by decade and director
# In each group (i.e., for each decade and each director),
# calculate the number of movies and average IMDB rating.
# Sort in descending order of movie count
imdb.groupby(['Decade', 'Director']).agg(
    Num_Movies=('Series_Title', 'count'),
    Avg_IMDB_Rating=('IMDB_Rating', 'sum')
).sort_values(by='Num_Movies', ascending=False)
```

```python
# Find the top 3 highest rated movies of each year since 2010.
# Show their released year, ranking, title, and IMDB_Rating.
# Sort them in descending order of year and ascending order of ranking
movies_since_2010 = imdb[imdb['Released_Year'] >= 2010].copy()

top_movies_by_year = movies_since_2010.sort_values(by=['Released_Year', 'IMDB_Rating'], ascending=[True, False])

top_movies_by_year = top_movies_by_year.groupby('Released_Year').head(3)

top_movies_by_year['Ranking'] = top_movies_by_year.groupby('Released_Year')['IMDB_Rating'].rank(ascending=False, method='first').astype(int)

top_three_by_year = top_movies_by_year[['Released_Year', 'Ranking', 'Series_Title', 'IMDB_Rating']].sort_values(
    by=['Released_Year', 'Ranking'], ascending=[False, True]
)

top_three_by_year
```

```python
# Find all directors whose movies appeared in at least five different decades.
# Your result should include: director, decade, and the number of movies in the decade.
decade_movie_counts = imdb.groupby(['Director', 'Decade']).agg(Num_Movies=('Series_Title', 'count')).reset_index()
multi_decade_directors = decade_movie_counts.groupby('Director')['Decade'].nunique()
multi_decade_directors = multi_decade_directors[multi_decade_directors >= 5].index.tolist()
result = decade_movie_counts[decade_movie_counts['Director'].isin(multi_decade_directors)]

result.sort_values(by=['Director', 'Decade'])
```
