### Tidying the data

Next, you will further tidy the two dataframes **stars** and **genres**.

Let's start with **stars**.

```python
# Transform the dataframe "stars" to a new dataframe named "stars_long" with the following four columns:
# Series_Title, Released_Year, StarNo (e.g., Star1, Star2, ...), StarName
# Hint: use melt()
stars_long = stars.melt(id_vars=['Series_Title', 'Released_Year'],
                       value_vars=['Star1', 'Star2', 'Star3', 'Star4'],
                       var_name='StarNo', value_name='StarName')
stars_long
```

```python
# Can you transform dataframe 'stars_long' back to its original shape?
# Hint: use pivot()
stars_long.pivot(index=['Series_Title', 'Released_Year'], columns='StarNo', values='StarName')
```

```python
# In dataframe "stars_long", find all movies that star "Morgan Freeman".
stars_long[stars_long['StarName'] == 'Morgan Freeman']
```

```python
# Who appeared in Star2 the most times? List the top five actors.
stars_long['StarName'][stars_long['StarNo'] == 'Star2'].value_counts().head(5)
```

```python
# Who stars in the most movies in this list? List the top 20 actors.
stars_long['StarName'].value_counts().head(20)
```

```python
# Which movie stars had the highest total gross in this movie list? Show the top 10 actors.
# Hint: Join "stars_long" and "imdb"; then group by StarName

(stars_long.merge(imdb[['Series_Title', 'Released_Year', 'Gross']], on=['Series_Title', 'Released_Year'], how='left')
 .groupby('StarName')['Gross'].sum()
 .nlargest(10))
```

```python
# Find the best director-actor duos
# i.e., director-actor pairs that collaborated in at least five movies,
# sort them in descending order of average IMDB_Rating
# Hint: Join imdb and stars_long; group by ['Director','StarName']

duo_collaborations = (
    imdb.merge(stars_long, on=['Series_Title', 'Released_Year'], how='inner')
    .groupby(['Director', 'StarName'])
    .agg(num_collaborations=('Series_Title', 'count'),
         avg_imdb_rating=('IMDB_Rating', 'mean'))
    .reset_index()
)

best_duos = (
    duo_collaborations[duo_collaborations['num_collaborations'] >= 5]
    .sort_values(by='avg_imdb_rating', ascending=False)
)

best_duos
```

```python
# Bonus question
# Who did "Amy Adams" co-star with in this movie list?
# Hint: Join stars_long with itself to find pairs of co-stars
amy_adams_movies = stars_long[stars_long['StarName'] == 'Amy Adams']
co_stars = amy_adams_movies.merge(stars_long,
                                  on=['Series_Title', 'Released_Year'],
                                  suffixes=('_adams', '_co_star'))
co_stars_list = co_stars[co_stars['StarName_co_star'] != 'Amy Adams']['StarName_co_star'].unique()

co_stars_list
```

Next, let's reshape the dataframe **genres**, which is a little bit more complicated.

```python
# Step 1: Split the 'Genre' string by ', ' into a list of individual genres and expand them to different columns
genres_split = genres['Genre'].str.split(', ', expand=True)
genres_split.head()
```

```python
# Step 2: Rename the columns as: Genre1, Genre2, Genre3
genres_split.columns = [f'Genre{i+1}' for i in range(genres_split.shape[1])]
genres_split.head()
```

```python
# Step 3: Combine ['Series_Title','Released_Year'] in 'genres' and ['Genre1','Genre2','Genre3'] in 'genres_split'.
# Save it to a new dataframe named 'genres_wide'.
# Hint: Use pd.concat(...)
genres_wide = pd.concat([genres[['Series_Title','Released_Year']], genres_split], axis=1)
genres_wide.head()
```

```python
# Step 4: Transform genres_wide to a new dataframe genres_long with the following four columns:
# Series_Title, Released_Year, GenreNo (e.g., Genre1, Genre2, Genre3), GenreName
# Hint: use melt()
genres_long = genres_wide.melt(id_vars=['Series_Title', 'Released_Year'],
                               value_vars=[col for col in genres_wide.columns if col.startswith('Genre')],
                               var_name='GenreNo', value_name='GenreName')
genres_long = genres_long.dropna(subset=['GenreName'])

genres_long.head()
```

```python
# How many unique genres (atomic values, e.g., Drama, Comedy, ...) are there?
# How many movies are there for each genre?
unique_genres_count = genres_long['GenreName'].nunique()
print(f"Unique genres: {unique_genres_count}")

genre_counts = genres_long['GenreName'].value_counts().reset_index()
genre_counts.columns = ['GenreName', 'Movie_Count']

genre_counts
```

```python
# What is the average IMDB rating for each genre?
# Sort the genres in descending order of average IMDB_Rating.
# Hint: join imdb with genres_long; group by GenreName
genre_imdb_ratings = imdb.merge(genres_long, on=['Series_Title', 'Released_Year'], how='inner')
avg_imdb_by_genre = genre_imdb_ratings.groupby('GenreName')['IMDB_Rating'].mean().reset_index()
avg_imdb_by_genre = avg_imdb_by_genre.sort_values(by='IMDB_Rating', ascending=False)

avg_imdb_by_genre
```

```python
# Who is the "King of Comedy" (i.e., the actor who starred in the most comedy movies)?
# Hint: find all comedies in genres_long ; join with stars_long; group by StarName
comedy_genres = genres_long[genres_long['GenreName'] == 'Comedy']
comedy_stars = comedy_genres.merge(stars_long, on=['Series_Title', 'Released_Year'], how='inner')
king_of_comedy = comedy_stars['StarName'].value_counts().reset_index()
king_of_comedy.columns = ['StarName', 'Comedy_Movie_Count']

king_of_comedy.head(1)
```

```python
# Who are the best action stars?
# i.e., the actors who star in at least 5 action movies, sorted by average IMDB rating in descending order
# Hint: You need to join columns from all three dataframes: imdb, stars_long, genres_long

# 1. Filter for action movies in genres_long
action_genres = genres_long[genres_long['GenreName'] == 'Action']

# 2. Join action_genres with stars_long to get actors in action movies
action_stars_df = action_genres.merge(stars_long, on=['Series_Title', 'Released_Year'], how='inner')

# 3. Join with imdb to get IMDB_Rating
action_stars_with_ratings = action_stars_df.merge(imdb[['Series_Title', 'Released_Year', 'IMDB_Rating']],
                                                    on=['Series_Title', 'Released_Year'],
                                                    how='inner')

# 4. Group by StarName to count movies and calculate average IMDB rating
star_action_stats = (
    action_stars_with_ratings.groupby('StarName')
    .agg(num_action_movies=('Series_Title', 'count'),
         avg_imdb_rating=('IMDB_Rating', 'mean'))
    .reset_index()
)

# 5. Filter for actors with at least 5 action movies
best_action_stars = star_action_stats[star_action_stats['num_action_movies'] >= 5]

# 6. Sort by average IMDB rating in descending order
best_action_stars = best_action_stars.sort_values(by='avg_imdb_rating', ascending=False)

display(best_action_stars)
```

```python
# Create a pivot table to show the number of movies of different genres in different decades.
# Row: GenreName
# Column: Decade
# Hint: https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html
genres_with_decade = genres_long.merge(imdb[['Series_Title', 'Released_Year', 'Decade']],
                                       on=['Series_Title', 'Released_Year'],
                                       how='inner')
pivot_table_genres_decades = pd.pivot_table(genres_with_decade,
                                            index='GenreName',
                                            columns='Decade',
                                            values='Series_Title',
                                            aggfunc='count')

pivot_table_genres_decades
```
