## DataFrame with columns

- year,month,day
        Date of departure    
- dep_time,arr_time
        Actual departure and arrival times (format HHMM or HMM), local tz.
- sched_dep_time,sched_arr_time
        Scheduled departure and arrival times (format HHMM or HMM), local tz.    
- dep_delay,arr_delay
        Departure and arrival delays, in minutes. Negative times represent early departures/arrivals.
- hour,minute
        Time of scheduled departure broken into hour and minutes.
- carrier
        Two letter carrier abbreviation. See airlines() to get name
- tailnum
        Plane tail number
- flight
        Flight number
- origin,dest
        Origin and destination. See airports() for additional metadata.
- air_time
        Amount of time spent in the air, in minutes
- distance
        Distance between airports, in miles
- time_hour
        Scheduled date and hour of the flight as a date. Along with origin, can be used to join flights data to weather data.

## Exercises

Write Python scripts to answer the following questions.

```python
# How many distinct airplanes are there in this dataset?
flights.tailnum.nunique()
```

```python
# Which hours of the day are the busiest?
# List the top 5 hours with the most flights.
flights.groupby('hour').size().sort_values(ascending=False).head(5)
```

```python
# Find all flights (identified by 'carrier' and 'flight')
# that always depart at least 60 minutes late.
late_flights = flights.groupby(['carrier', 'flight'])['dep_delay'].min()
late_flights = late_flights[late_flights >= 60]
late_flights
```

```python
# How many flights always depart at least 60 minutes late?
print(f'{len(late_flights)} flights always depart at least 60 minutes late')
```

```python
# Find all United Airlines (UA) flights (identified by 'carrier' and 'flight')
# that depart (at least) 60 minutes late.
UA_late_flights = flights[
    (flights.carrier == 'UA') &
    (flights.dep_delay >= 60)
    ]
UA_late_flights = UA_late_flights.groupby(['carrier', 'flight']).nunique()

UA_late_flights
```

```python
# How many United Airlines (UA) flights depart (at least) 60 minutes late?
print(f'{len(UA_late_flights)} United Airlines (UA) flights depart (at least) 60 minutes late')
```

```python
# Which flights that always arrived on time in December?
december_flights = flights[flights.month == 12]
nlate_december_flights = december_flights.groupby(['carrier', 'flight'])['arr_delay'].max()
nlate_december_flights = nlate_december_flights[nlate_december_flights <= 0]

nlate_december_flights
```

```python
# How many flights always arrived on time in December?
print(f'{len(nlate_december_flights)} flights always arrived on time in December')
```

```python
# Create a dataframe 'cancelled' for all cancelled flights.
# Assumption: flights that are not cancelled should have values for dep_time.
cancelled_flights = flights[flights.dep_time.isnull()]

cancelled_flights
```

```python
# Look at the number of cancelled flights per day.
# Find the top 10 days with the most cancelled flights.
top_days_cancelled_flights = cancelled_flights.groupby(['year', 'month', 'day']).size().sort_values(ascending=False)
top_days_cancelled_flights.head(10)
```

```python
# Which carrier has the worst average arrival delay?
worst_carrier_arr_delay = flights.groupby('carrier')['arr_delay'].mean().sort_values(ascending=False)

print(f'{worst_carrier_arr_delay.index[0]} has the worst average arrival delay')
```

```python
# Which plane (tailnum) has the worst on-time record?
worst_on_time_planes = flights.groupby('tailnum')['arr_delay'].mean().sort_values(ascending=False)

print(f'{worst_on_time_planes.index[0]} has the worst on-time record')
```

```python
# What time (hour) of day should you fly if you want to avoid delays as much as possible?
min_delays_flights = flights.groupby('hour')[['arr_delay', 'dep_delay']].mean()
min_delays_flights['total_avg_delay'] = min_delays_flights.arr_delay + min_delays_flights.dep_delay
min_delays_flights = min_delays_flights['total_avg_delay'].sort_values(ascending=True)

print(f'fly at {min_delays_flights.index[0]} to avoid delays as much as possible')
```

```python
# For each destination, compute the total minutes of arrival delay.
total_arr_delay = flights.groupby('dest')['arr_delay'].sum()

total_arr_delay
```

```python
# Find the worst days in terms of average arrival delay greater than 1 hour.
worst_days = flights.groupby(['year', 'month', 'day'])['arr_delay'].mean()
worst_days = worst_days[worst_days > 60].sort_values(ascending=False)

worst_days
```

```python
# Find the number of carriers between each pair of origin and destination.
carrier_count = flights.groupby(['origin', 'dest'])['carrier'].nunique()

carrier_count
```
